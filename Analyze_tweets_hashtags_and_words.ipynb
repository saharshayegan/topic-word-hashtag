```python
from collections import Counter
import pandas as pd
import hazm as hz
```

```python
topic = 'sports'
result = pd.read_csv('./drive/MyDrive/Colab Notebooks/topic_words/'+topic+'_tweets.csv')
```

```python
import numpy as np

hash_dict = {}
for i,row in result.iterrows():
  if type(row['Hashtags']) != float:
    tokens = ((row['Hashtags'].replace('[',',')).replace(']', ',')).split(',') 
    for t in tokens:
      hash_dict[t] = 0
for i,row in result.iterrows():
  if type(row['Hashtags']) != float:
    tokens = ((row['Hashtags'].replace('[',',')).replace(']', ',')).split(',') 
    for t in tokens:
      # print(t)
      hash_dict[t] = hash_dict[t] + 1

new_dict=pd.DataFrame((sorted(hash_dict.items(), key=lambda item: item[1])))
new_dict.to_csv('./drive/MyDrive/Colab Notebooks/topic_words/'+topic+'_hashtags.csv',index=False)
```
